## Here I am going to perform analysis over cancer death data. We will perform linear regression analysis over two column year and death. 

### As I suspect that cancer death has increased over the year.

```python
import pandas as pd;
#Get CSV data
cancer_death_data = pd.read_csv('./total-cancer-deaths-by-type.csv')
cancer_death_data.head(50)
```

### I am going to perform analysis over two column Year and deaths liver cancer.

### Here Year is independent variable and deaths is dependent variable

```python
import numpy as np

year = np.array(cancer_death_data.Year)
deaths = np.array(cancer_death_data['Deaths - Liver cancer - Sex: Both - Age: All Ages (Number)'])

```

### Plot Scatter chart to show the relation between Year and Death

```python
#Plot create function
import matplotlib.pyplot as plt
def show_scatter_plot(title, 
                      x_label, y_label, 
                      x_values, y_values):
    plt.rcParams.update({'font.size': 22})
    plt.figure(figsize=(26, 7)) # resize the chart
    plt.scatter(x_values, y_values)
    plt.title(title)
    plt.ylabel(y_label)
    plt.xlabel(x_label)
    plt.show()
```

```python
show_scatter_plot('Scatter Plot', 'Year Independent Variable', 
                  'Death Dependent Variable', year, deaths)
```

### Draw the least-squares regression line

```python
def calculate_slope_intercept(year, deaths):
    x = year
    y = deaths
    
    n = len(x)
    sum_x  = np.sum(x)
    sum_y  = np.sum(y)
    sum_xx = np.sum(x*x)
    sum_xy = np.sum(x*y)
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    
    num   = sum_xy - (sum_x*sum_y)/n
    den = sum_xx - ((sum_x*sum_x)/n)
    
    m = num/den
    b = mean_y - m*mean_x
    
    return m, b  # slope and intercept
```

```python
#Calculate regression line
def calculate_regression_line(x_values, y_values):
    m, b = calculate_slope_intercept(x_values, y_values)
    reg_line = [m*x + b for x in x_values]  # regression line
    return (reg_line, m, b)
```

### Show least square line on scatter chart 

```python
def show_least_squares_line(title, x_label, y_label, 
                            x_values, y_values):
    plt.figure(figsize=(26, 7)) # resize the chart
    plt.rcParams.update({'font.size': 22})
    plt.scatter(x_values, y_values)
    
    # Now show the least squares line.
    reg_line, m, b = calculate_regression_line(x_values, y_values)
    plt.plot(x_values, reg_line, color='red')
    plt.title(f'{title}, m = {m:.2f}, b = {b:.2f}')
    plt.ylabel(y_label)
    plt.xlabel(x_label)
    plt.show()
```

```python
show_least_squares_line('Linear Regression Line', 'Year Independent variable', 
                        'Deaths Dependent variable', year, deaths)
```

### As we can see in chart, values are not much scattered from linear regression line now we will calculate the Coefficient and Correlation to analyze the data further.

### Coefficient of Determination

```python
def calculate_coefficient_determination(x_values, y_values):
    y_avg = np.average(y_values)
    reg_line, m, b = calculate_regression_line(x_values, y_values)
    s_total = np.sum(pow((y_values-y_avg),2))
    s_residual = np.sum(pow((y_values-reg_line), 2))
    
    coefficient = 1-s_residual/s_total
    print('s_residual=%4.3f' %s_residual)
    print("s_total=%4.3f" %s_total)
    print("Coefficient of Determination= %2.3f" %coefficient)
```

```python
#Get the Coefficient determination
calculate_coefficient_determination(year, deaths)
```

### As per the calculated Coefficient determination = 0.948 it is very near value to 1 hence, it shows how data is not scattered.

### Correlation Coefficient

```python
import math
def calculate_correlation_coefficient(x_values, y_values):
        y_avg = np.average(y_values)
        x_avg = np.average(x_values)
        
        num = np.sum((x_values-x_avg)*(y_values-y_avg))
        den1 = math.sqrt(np.sum(pow((x_values-x_avg),2)))
        den2 = math.sqrt(np.sum(pow((y_values-y_avg),2)))
        den = den1 * den2
        correlation = num/den
        print("Correlation coefficient= %2.3f" % correlation)
    
```

```python
#Get the Correlation of coefficient
calculate_correlation_coefficient(year, deaths)
```

### As per the calculated Correlation coefficient= 0.974 it shows the how Year and deaths data related to one another.
