### Here I am going to use diabetes Age dataset to perform central limit theorem of statistics

```python
import pandas as pd

diabetes_data_set = pd.read_csv('diabetes.csv')
data = diabetes_data_set[['Age', 'Pregnancies']]
df = pd.DataFrame(data)

df.head(10)
```

```python
df['Age'].describe()
```

```python
import numpy as np

##function to calculate IQR
def iqr(data): 
    ages    = data
    count   = len(ages)
    min_age = min(ages)
    max_age = max(ages)
    
    q1  = np.percentile(ages, 25)
    q2  = np.percentile(ages, 50)
    q3  = np.percentile(ages, 75)
    iqr = q3 - q1
    return iqr
```

### Calculate three population parameters of age dataset of diabetes eg.(mean, median, IQR)

```python
#Mean of age data set is

import statistics as stat

mean_of_age = stat.mean(df['Age'])

#Median of age data set is
median_of_age = stat.median(df['Age'])

#standard deviation of age data is
std = stat.pstdev(df['Age'])


print('Mean of data set is', mean_of_age)
print('Median of data set is', median_of_age)
print('Interquartile range of data set is', iqr(df['Age']))
print('Standard deviation of age data', std)
```

### Random sampling of data using python random.sample function


```python
import random #declare a generic function random sampling of data
age_data = (df.Age).tolist(); #convert in list

# Write generic function to do sampling of data
def random_sampling(count, size):
    sample_mean = [] #Initialize the empty array
    for i in range(count):
        sample_list = random.sample(age_data, size) #Get random sample from given list
        sample_mean.append(stat.mean(sample_list)) #Get mean sample data
    return sample_mean

#Write a generic function to draw a graph for further analysis
def draw_chart(sample):
    #Draw a chart
    x1, y1 = np.unique(sample, return_counts = 'true') # get unique value and frequency
    plt.figure(figsize=(30, 10)) # resize the bar chart
    plt.rcParams['font.family'] = 'sans-serif' # style the font
    plt.rcParams['axes.linewidth']=1
    plt.rcParams['axes.labelsize'] = 20 # resize the label text
    plt.ylabel('Frequency')
    plt.xlabel('Age')
    bar = plt.bar(x1, y1, align='edge', width=0.3)
    
```
